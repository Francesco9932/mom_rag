```python
# Document Loader and text splitter imports
from langchain.document_loaders import HuggingFaceDatasetLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
# LLM Model import:
import torch
from transformers import BitsAndBytesConfig, AutoModelForCausalLM, AutoTokenizer, GenerationConfig, pipeline
from langchain import HuggingFacePipeline
# Langchain Pipeline and Template import
from langchain.chains.summarize import load_summarize_chain
from langchain_core.prompts import PromptTemplate
```

```python
text = "Tutti gli altri. Estate. Fantastico. Quindi, passiamo avanti e prendiamo il numero 16. Comunicazione della Consigliera Price raccomandazione di riferire alla Commissione Parchi e Ricreazione per considerare di nominare il campo sportivo. Il Campo Sportivo presso il Parco Marina Vista. Il Campo Luke to Sue Johnson. Consigliera Price. Grazie. Vice Sindaco, voglio... So che questo è un argomento che i miei colleghi supporteranno, ma non posso resistere a parlare un po' del motivo per cui è stato proposto, perché questa persona è stata molto speciale nella mia vita e nella vita della mia comunità. Nel 2005, un gruppo di studenti del terzo distretto, ragazzi delle scuole medie e superiori, hanno partecipato a un programma di bilancio partecipativo giovanile. Un membro importante del comitato studentesco era Luke Johnson, quindicenne. Luke era residente ad Alamitos Heights e viveva vicino al parco con sua madre, Rena, suo padre Todd e suo fratello Spencer. Luke fu diagnosticato leucemia all'età di 14 anni, proprio il giorno in cui iniziò il suo primo anno di liceo alla Wilson High School. Nel 2006, Luke perse la sua battaglia di due anni con la leucemia. Questo percorso lo ha allontanato dalla scuola per gran parte della sua carriera scolastica iniziale. Il comitato divenne uno dei modi in cui riusciva a rimanere in contatto con i suoi coetanei durante la sua malattia ed era una responsabilità che prendeva seriamente. Luke era molto amato nella comunità e era amico di tutti. È profondamente mancato dalla comunità, dai suoi coetanei e dalla sua famiglia. In onore del suo coinvolgimento nella comunità e del ruolo che ha avuto nella costruzione del Campo Sportivo Marina Vista, la comunità ha espresso il suo sostegno per nominare il nuovo campo sportivo in memoria di Luke Johnson. Chiedo ai miei colleghi di sostenere questa proposta. Ho chiesto ai genitori di Luke di non venire stasera perché so che questo dovrà passare attraverso un processo e non voglio che debbano continuare a venire alle riunioni del consiglio. Una volta che il processo sarà completato, faremo una nomina ufficiale al parco e organizzeremo una bella celebrazione con la sua famiglia e i membri della nostra comunità. Grazie, Consigliera. Consigliere Austin, commento pubblico sullo stesso. Ciao. Mi chiamo Alex Phillips e Luke Johnson è un ragazzo che vive nel mio quartiere. Conosco i suoi genitori, conosco i suoi fratelli. E so che questo significherebbe molto per l'intera comunità. E penso che sia meraviglioso che possiamo fare qualcosa del genere perché è stata una grande tragedia. E grazie, Susie, per averlo portato. Grazie. Grazie. E nessun ulteriore commento pubblico. E voglio solo dire, Consigliere, che penso che questa sia un'ottima idea. Sono felice di votare e sostenere questo stasera. E membri, per favore votate. La mozione è stata approvata. Grazie. Numero 15, per favore."

splitter = RecursiveCharacterTextSplitter()
from langchain.docstore.document import Document
docs = splitter.split_text(text)
docs = [Document(page_content=t) for t in docs]
```

```python
MODEL_NAME = "galatolo/cerbero-7b"

quantization_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.float16,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_use_double_quant=True,
)

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME, use_fast=True)
tokenizer.pad_token = tokenizer.eos_token

# Senza quantizzazione servono 15GB di VRAM

model = AutoModelForCausalLM.from_pretrained(
    MODEL_NAME,
    torch_dtype=torch.float16,
    trust_remote_code=True,
    device_map="auto",
    # offload_folder="offload"
    quantization_config=quantization_config
)

generation_config = GenerationConfig.from_pretrained(MODEL_NAME)
generation_config.max_new_tokens = 1024
generation_config.temperature = 0.0001
generation_config.top_p = 0.95
generation_config.do_sample = True
generation_config.repetition_penalty = 1.15

pipeline = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    return_full_text=True,
    generation_config=generation_config,
)

llm = HuggingFacePipeline(
    pipeline=pipeline,
    )
```

```python
prompt_template = """
Sei un assistente AI specializzato nel riassumere meeting in maniera precisa.


{text}


RIASSUNTO COINCISO USANDO BULLET POINTS:
"""

prompt = PromptTemplate(template=prompt_template, input_variables=["text"])

chain = load_summarize_chain(llm, chain_type="stuff", prompt=prompt)

# label: Recommendation to approve renaming the Multi-Use Sport Court in Marina Vista Park the Luke Tatsu Johnson Court.

print(chain.run(docs))
```
